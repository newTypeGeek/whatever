```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

\begin{align}
    x(t) \; = \;  \big(x_{\infty} - x_{0}\big) \, \big( 1 - e^{-t/\tau}) \, + \, x_{0} \, + \, \eta(t)
\end{align}

```python
n = 60*4
x_inf = 5.0
x_0 = 10.0
tau = 30

noise_mean = 0.
noise_std = 0.3

index = pd.date_range("2020-01-01", freq="1T", periods=n, name="timestamp")

t = np.arange(0, n)
x = (x_inf - x_0) * (1 - np.exp(-t/tau)) + x_0  + np.random.normal(noise_mean, noise_std, t.shape)


df = pd.DataFrame(data={"x": x}, index=index)
df
```

```python
fig, ax = plt.subplots()
df.plot(ax=ax)

x_lim = ax.get_xlim()
ax.hlines(xmin=x_lim[0], xmax=x_lim[1], y=x_inf, color="red", linestyle="--")

ax.set_ylabel("x(t)")


```
